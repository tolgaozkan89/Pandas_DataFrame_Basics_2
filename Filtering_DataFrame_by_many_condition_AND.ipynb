# PART1: "titanic" DATASET

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head(10)
```

```python
titanic.columns
```

## mask1 = "male"

```python
mask1 = titanic.sex == "male"
mask1
```

## mask2 = "age" > 14

```python
mask2 = titanic.age > 14
mask2.head()
```

```python
( mask1 & mask2 ).head()
```

```python
male_surv = titanic.loc[mask1 & mask2, ["survived","pclass","sex","age"]]
male_surv.head(20)
```
