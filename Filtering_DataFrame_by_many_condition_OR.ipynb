# PART1: "titanic" DATASET

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head(10)
```

```python
titanic.columns
```

## mask1 = "female"

```python
mask1 = titanic.sex == "female"
mask1
```

## mask2 = "age" < 14

```python
mask2 = titanic.age < 14
mask2.head()
```

```python
( mask1 | mask2 ).head()
```

```python
titanic.loc[mask1 | mask2]
```

```python
wom_or_chi = titanic.loc[mask1 | mask2, ["survived","pclass","sex","age"]]
wom_or_chi.head()
```

```python
wom_or_chi.info()
```

```python
wom_or_chi.describe()
```
