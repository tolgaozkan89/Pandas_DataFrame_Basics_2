# between()

## between( Left, Right, Boundary Inclusive = true or false)

```python
summer["Year"].between(1960,1969,inclusive=True)
```

```python
og_60s = summer.loc[summer["Year"].between(1960,1969,inclusive=True)]
og_60s
```

```python
og_60s.head()
```

```python
og_60s.tail()
```

# isin()

```python
summer["Year"].isin([1972])
```

```python
summer["Year"].isin([1972,1996])
```

```python
og_72_96 = summer.loc[summer["Year"].isin([1972,1996])]
og_72_96
```

```python
my_favorite_games = [1972,1996]
```

```python
og_72_96 = summer.loc[summer["Year"].isin(my_favorite_games)]
```

```python
og_72_96.head()
```

```python
og_72_96.tail()
```

# ~ directly makes the opposite of isin(), like NOT

```python
og_not_72_96 = summer.loc[~summer["Year"].isin(my_favorite_games)]
```

```python
og_not_72_96.head()
```

```python
og_not_72_96.tail()
```

Below you can see all years except 1972 and 1996

```python
og_not_72_96["Year"].unique()
```

```python
og_not_72_96["Year"].nunique()
```
